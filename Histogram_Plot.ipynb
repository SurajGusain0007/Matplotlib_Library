#Histogram

**t*Use histograms when you have continuous measurements and want to understand the distribution of values and look for outliers. These graphs take your continuous measurements and place them into ranges of values known as bins.The histogram is a popular graphing tool. It is used to summarize discrete or continuous data that are measured on an interval scale. It is often used to illustrate the major features of the distribution of the data in a convenient form**


```python
import matplotlib.pyplot as plt
import numpy as np
mi_students=np.random.randint(18,45,(100))
py_students=np.random.randint(15,40,(100))
```

```python
plt.hist(mi_students)
plt.title("mi_students age")
plt.xlabel("Students age category")
plt.ylabel("No of students age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(mi_students,bins,rwidth=0.8)
plt.title("mi_students age")
plt.xlabel("Students age category")
plt.ylabel("No of students age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(mi_students,bins,rwidth=0.8,histtype='step')
plt.title("mi_students age")
plt.xlabel("Students age category")
plt.ylabel("No of students age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(mi_students,bins,rwidth=0.8,orientation='horizontal')
plt.title("mi_students age")
plt.xlabel("Students age category")
plt.ylabel("No of students age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(mi_students,bins,rwidth=0.8,color='m')
plt.title("mi_students age")
plt.xlabel("Students age category")
plt.ylabel("No of students age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(mi_students,bins,rwidth=0.8,label='Ml_students')
plt.title("mi_students age")
plt.xlabel("Students age category")
plt.ylabel("No of students age")
plt.legend(loc=1)
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(mi_students,bins,rwidth=0.8,label='Ml_students')
plt.hist(py_students,bins,rwidth=0.8,label='py_students')
plt.title("mi_students age vs Py_Students age")
plt.xlabel("Students age category")
plt.ylabel("No of students age")
plt.legend(loc=1)
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.figure(figsize=(16,9))
plt.hist([mi_students,py_students],bins,rwidth=0.8,label=['Ml_students','py_students'],color=['r','g'])
plt.title("mi_students age vs Py_Students age")
plt.xlabel("Students age category")
plt.ylabel("No of students age")
plt.legend(loc=1)
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
bins=[15,20,25,30,35,40,45]
plt.figure(figsize=(16,9))
plt.hist([mi_students,py_students],bins,rwidth=0.8,label=['Ml_students','py_students'],color=['r','g'])
plt.title("mi_students age vs Py_Students age")
plt.xlabel("Students age category")
plt.ylabel("No of students age")
plt.legend(loc=1)
plt.grid(color='k',linestyle='--')
plt.show()
```

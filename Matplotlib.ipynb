#BAR GRAPH

**If you have comparative data that you would like to represent through a chart then a bar chart would be the best option. This type of chart is one of the more familiar options as it is easy to interpret. These charts are useful for displaying data that is classified into nominal or odinal categories.A bar chart uses bars to show comparisons between categories of data. These bars can be displayed horizontally or vertically. A bar graph will always have two axis. One axis will generally have numerical values, and the other will describe the types of categories being compared.**

```python
import matplotlib.pyplot as plt
x=['Python','C','C++','Java']
y=[85,70,60,82]
c=['y','b','g','m']
plt.bar(x,y,width=0.4,color=c,edgecolor='r',linewidth=5,linestyle=":",alpha=0.4)
plt.xlabel("Language",fontsize=12)
plt.ylabel("No.",fontsize=12)
plt.title("Popularity")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['Python','C','C++','Java']
y=[85,70,60,82]
c=['y','b','g','m']
plt.bar(x,y,width=0.4,color=c,label='popularity')
plt.xlabel("Language",fontsize=12)
plt.ylabel("No.",fontsize=12)
plt.title("Popularity")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=['Python','C','C++','Java']
y=[85,70,60,82]
z=[50,65,45,70]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
plt.bar(p,y,width,color='r')
plt.bar(p1,z,width,color='y')
plt.xlabel("Language",fontsize=12)
plt.ylabel("No.",fontsize=12)
plt.title("Popularity")
plt.xticks(p+width/2,x,rotation=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=['Python','C','C++','Java']
y=[85,70,60,82]
z=[50,65,45,70]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]
plt.barh(p,y,width,color='r')
plt.barh(p1,z,width,color='y')
plt.xlabel("Language",fontsize=12)
plt.ylabel("No.",fontsize=12)
plt.title("Popularity")
plt.show()
```

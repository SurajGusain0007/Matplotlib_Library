#Scatter Plot

**A scatter chart works best when comparing large numbers of data points without regard to time. This is a very powerful type of chart and good when your are trying to show the relationship between two variables (x and y axis), for example a person's weight and height. A good example of this can be seen below.Scatter plot graph is used to study the relationship between two variables. It basically shows how one variable is affected by the other.
Scatter plots usually consist of a large body of data. The closer the data points come when plotted to making a straight line, the higher the correlation between the two variables, or the stronger the relationship.
If the data points make a straight line going from the origin out to high x- and y-values, then the variables are said to have a positive correlation . If the line goes from a high-value on the y-axis down to a high-value on the x-axis, the variables have a negative correlation .**



**

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
color=['r','g','y','r','g','r','y']
size=[100,200,110,130,400,200,150]
plt.scatter(x,y,c=color,s=size,alpha=0.5)
plt.title("Scatterplot",fontsize=10)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
color=['r','g','y','r','g','r','y']
size=[100,200,110,130,400,200,150]
plt.scatter(x,y,c=color,s=size,alpha=0.5,marker="*",edgecolor='g',linewidth=1)
plt.title("Scatterplot",fontsize=10)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
color=['r','g','y','r','g','r','y']
size=[100,200,110,130,400,200,150]
plt.scatter(x,y,cmap='viridis',s=size)
t=plt.colorbar()
t.set_label("colorbar",fontsize=15)
plt.title("Scatterplot",fontsize=10)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
z=[4,5,6,1,3,7,2]
size=[100,200,110,130,400,200,150]
plt.scatter(x,y,c='r',s=size,marker="*")
plt.scatter(x,z,c='g',s=size)
plt.title("Scatterplot",fontsize=10)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

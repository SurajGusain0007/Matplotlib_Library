#Line Plot

**Line Plots display numerical values on one axis, and categorical values on the other. They can typically be used in much the same way Bar Plots can be used, though, they're more commonly used to keep track of changes over time.**

```python
import matplotlib.pyplot as plt
y = [1, 5, 3, , 7, 8]
x = [10, 20, 30, 40, 50, 60]
plt.plot(x,y)
plt.xlabel("x",fontsize=16)
plt.ylabel("y",fontsize=16)
plt.title("Display",fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
y = [1, 5, 3, 10, 7, 8]
x = [10, 20, 30, 40, 50, 60]
plt.plot(x,y)
plt.axis([0,70,0,15])
plt.xlabel("x",fontsize=16)
plt.ylabel("y",fontsize=16)
plt.title("Display",fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
y = [1, 5, 3, 10, 7, 8]
x = [10, 20, 30, 40, 50, 60]
plt.plot(x,y,marker='o',linestyle='dashed',color='r',linewidth=2,markersize=6)
plt.axis([0,70,0,15])
plt.xlabel("x",fontsize=16)
plt.ylabel("y",fontsize=16)
plt.title("Display",fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
y = [1, 5, 3, 10, 7, 8]
x = [10, 20, 30, 40, 50, 60]
plt.plot(x,y,'go--',linewidth=2,markersize=6)
plt.axis([0,70,0,15])
plt.xlabel("x",fontsize=16)
plt.ylabel("y",fontsize=16)
plt.title("Display",fontsize=16)
plt.legend(['dis_line'],loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
y = [1, 5, 3, 10, 7, 8]
x = [10, 20, 30, 40, 50, 60]
style.use("ggplot")
plt.plot(x,y,'go--',linewidth=2,markersize=6,label='dis_line')
plt.axis([0,70,0,15])
plt.xlabel("x",fontsize=16)
plt.ylabel("y",fontsize=16)
plt.title("Display",fontsize=16)
plt.grid(color='k',linestyle='-')
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
y = [1, 5, 3, 10, 7, 8]
x = [10, 20, 30, 40, 50, 60]
z=  [15,25,35,45,55,65]
style.use("ggplot")
plt.plot(x,y,'go--',linewidth=2,markersize=6,linestyle='dashed',label='MiniDisplay')
plt.plot(z,y,'ro--',linewidth=2,markersize=6,linestyle='dotted',label='Display')
plt.axis([0,70,0,15])
plt.xlabel("x",fontsize=16)
plt.ylabel("y",fontsize=16)
plt.title("Mini Display and Display",fontsize=16)
plt.grid(color='k',linestyle='-')
plt.legend(loc=2)
plt.show()
```
